```python
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors
import numpy as np

# Data
models = [
    "Qwen-2 7B Instruct", "Gemma 7B Instruct", "Mistral 7B Instruct V0.2", 
    "Phi-3 Medium 4K Instruct", "Medical LLaMA-3 8B", "LLaMA-3 8B Instruct"
]

pretrained_data = {
    "Qwen-2 7B Instruct": {"accuracy": 0.252, "f1-score": 0.248},
    "Gemma 7B Instruct": {"accuracy": 0.191, "f1-score": 0.190},
    "Mistral 7B Instruct V0.2": {"accuracy": 0.223, "f1-score": 0.226},
    "Phi-3 Medium 4K Instruct": {"accuracy": 0.272, "f1-score": 0.262},
    "Medical LLaMA-3 8B": {"accuracy": 0.172, "f1-score": 0.191},
    "LLaMA-3 8B Instruct": {"accuracy": 0.267, "f1-score": 0.251},
}

fine_tuned_data = {
    "Qwen-2 7B Instruct": {"accuracy": 0.843, "f1-score": 0.848},
    "Gemma 7B Instruct": {"accuracy": 0.657, "f1-score": 0.659},
    "Mistral 7B Instruct V0.2": {"accuracy": 0.735, "f1-score": 0.737},
    "Phi-3 Medium 4K Instruct": {"accuracy": 0.673, "f1-score": 0.682},
    "Medical LLaMA-3 8B": {"accuracy": 0.756, "f1-score": 0.766},
    "LLaMA-3 8B Instruct": {"accuracy": 0.968, "f1-score": 0.961},
}
# Calculate improvement score for sorting
improvement_scores = {model: fine_tuned_data[model]["accuracy"] - pretrained_data[model]["accuracy"] for model in models}

# Sort models by improvement score
sorted_models = sorted(models, key=lambda x: improvement_scores[x], reverse=True)

# Create a color gradient from dark blue to red
colors = plt.cm.plasma(np.linspace(0.1, 0.9, len(models)))

# Create a color map based on sorted models
model_colors = {model: colors[i] for i, model in enumerate(sorted_models)}

plt.figure(figsize=(10, 6))

# Plotting data points and lines
for model in sorted_models:
    pre_acc = pretrained_data[model]["accuracy"]
    pre_f1 = pretrained_data[model]["f1-score"]
    fine_acc = fine_tuned_data[model]["accuracy"]
    fine_f1 = fine_tuned_data[model]["f1-score"]

    # Plot pre-trained and fine-tuned points
    plt.scatter(pre_acc, pre_f1, color=model_colors[model], label=f"{model} (Pre-trained)")
    plt.scatter(fine_acc, fine_f1, color=model_colors[model], marker='x', label=f"{model} (Fine-tuned)")

    # Draw line between pre-trained and fine-tuned points
    plt.plot([pre_acc, fine_acc], [pre_f1, fine_f1], color=model_colors[model], linestyle='--')

# Adding labels and legend
plt.xlabel('Accuracy')
plt.ylabel('F1-Score')
plt.title('Accuracy vs F1-Score for Pre-trained and Fine-tuned Models')
plt.legend(loc='upper left', bbox_to_anchor=(1, 1))
plt.grid(True)

# Save the plot as PNG
plt.tight_layout()
plt.savefig('accuracy_vs_f1_score.png', dpi=300)
plt.show()
```

```python
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors
import numpy as np

# Data
models = [
    "Qwen-2 7B Instruct", "Gemma 7B Instruct", "Mistral 7B Instruct V0.2", 
    "Phi-3 Medium 4K Instruct", "Medical LLaMA-3 8B", "LLaMA-3 8B Instruct"
]

pretrained_data = {
    "Qwen-2 7B Instruct": {"precision": 0.408, "recall": 0.252},
    "Gemma 7B Instruct": {"precision": 0.254, "recall": 0.191},
    "Mistral 7B Instruct V0.2": {"precision": 0.367, "recall": 0.223},
    "Phi-3 Medium 4K Instruct": {"precision": 0.404, "recall": 0.272},
    "Medical LLaMA-3 8B": {"precision": 0.378, "recall": 0.172},
    "LLaMA-3 8B Instruct": {"precision": 0.366, "recall": 0.267},
}

fine_tuned_data = {
    "Qwen-2 7B Instruct": {"precision": 0.916, "recall": 0.843},
    "Gemma 7B Instruct": {"precision": 0.725, "recall": 0.657},
    "Mistral 7B Instruct V0.2": {"precision": 0.787, "recall": 0.735},
    "Phi-3 Medium 4K Instruct": {"precision": 0.783, "recall": 0.673},
    "Medical LLaMA-3 8B": {"precision": 0.863, "recall": 0.756},
    "LLaMA-3 8B Instruct": {"precision": 0.957, "recall": 0.968},
}

# Calculate improvement score for sorting
improvement_scores = {model: fine_tuned_data[model]["precision"] - pretrained_data[model]["precision"] for model in models}

# Sort models by improvement score
sorted_models = sorted(models, key=lambda x: improvement_scores[x], reverse=True)

# Create a color gradient from dark blue to red
colors = plt.cm.plasma(np.linspace(0.1, 0.9, len(models)))

# Create a color map based on sorted models
model_colors = {model: colors[i] for i, model in enumerate(sorted_models)}

plt.figure(figsize=(10, 6))

# Plotting data points and lines
for model in sorted_models:
    pre_precision = pretrained_data[model]["precision"]
    pre_recall = pretrained_data[model]["recall"]
    fine_precision = fine_tuned_data[model]["precision"]
    fine_recall = fine_tuned_data[model]["recall"]

    # Plot pre-trained and fine-tuned points
    plt.scatter(pre_precision, pre_recall, color=model_colors[model], label=f"{model} (Pre-trained)")
    plt.scatter(fine_precision, fine_recall, color=model_colors[model], marker='x', label=f"{model} (Fine-tuned)")

    # Draw line between pre-trained and fine-tuned points
    plt.plot([pre_precision, fine_precision], [pre_recall, fine_recall], color=model_colors[model], linestyle='--')

# Adding labels and legend
plt.xlabel('Precision')
plt.ylabel('Recall')
plt.title('Precision vs Recall for Pre-trained and Fine-tuned Models')
plt.legend(loc='upper left', bbox_to_anchor=(1, 1))
plt.grid(True)

# Save the plot as PNG
plt.tight_layout()
plt.savefig('precision_vs_recall.png', dpi=300)
plt.show()
```
